```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#First dates
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Linear_regration_one_demetion_xy_format_100.csv', sep=",")
data
```

```python
X = data['Free'].values.reshape(-1, 1)
y= data['Depended'].values
model = LinearRegression()
model.fit(X,y)

predict_y = model.predict(X)
```

```python
#Строим график
plt.scatter(X, y, color='blue', label='Данные')
plt.plot(X, predict_y, color='red', label='Линейная регрессия')
plt.xlabel('X')
plt.ylabel(' y')
plt.title('Простая линейная регрессия')
plt.legend()
plt.show()
```
